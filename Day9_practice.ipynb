```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.drop(columns='Loan_ID', axis=0, inplace=True)
df.head()
```

```python
df['Gender'] = df['Gender'].map({"Male":0,"Female":1})
df['Married'] = df['Married'].map({"Yes":0,"No":1})
df['Dependents'] = df['Dependents'].map({"0":0,"1":1,"2":2,"3+":3})
df['Education'] = df['Education'].map({"Graduate":0,"Not Graduate":1})
df['Self_Employed'] = df['Self_Employed'].map({"Yes":0,"No":1})
df['Property_Area'] = df['Property_Area'].map({"Rural":0,"Urban":1,"Semiurban":2})
df['Loan_Status'] = df['Loan_Status'].map({"Y":0,"N":1})
df
```

```python
X = df.iloc[:,df.columns!='Loan_Status']
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
pred_value = lr.predict(X_test)
```

```python
import pickle
with open('testing_day_9.pkl','wb') as f:
    pickle.dump(lr, f)
```

```python
import joblib

model = joblib.load('testing_day_9.pkl')
score = model.predict(X_test)
score
```
